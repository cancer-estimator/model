```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/Users/oscar/PycharmProjects/PO235/Dados Lung Cancer Data Set - Manoel/survey lung cancer.csv")
df.head()
```

```python
df.drop_duplicates(inplace=True)
# exclui registros duplicados
df.describe()
```

```python
df.shape
df.isnull().values.any()
# verifica se existem valores nulos no dataset
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['LUNG_CANCER']=encoder.fit_transform(df['LUNG_CANCER'])
df['GENDER']=encoder.fit_transform(df['GENDER'])
df.head()
df.info()
# transforma as variáveis gender e lung cancer em números M vira 1 e YES vira 1.
```

```python
con_col = ['AGE']
cat_col=[]
for i in df.columns:
    if i!='AGE':
        cat_col.append(i)
# separa as colunas em uma coluna de controle e outra de categoria
```

```python
X=df.drop(['LUNG_CANCER'],axis=1)
y=df['LUNG_CANCER']
# particiona o dataframe em X todas as colunas menos o lung cancer e y só a coluna do LUNG CANCER (Target)

for i in X.columns[2:]:
    temp=[]
    for j in X[i]:
        temp.append(j-1)
    X[i]=temp
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size = 0.2)
# divide o data set em um conjunto de treino e outro de teste (validação)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
def custom_predict(X, threshold):
    probs = model.predict_proba(X) 
    return (probs[:, 1] > threshold).astype(int)

new_preds = custom_predict(X=X_test, threshold=0.3)
accuracy = accuracy_score(y_test, new_preds)
print("Accuracy:", accuracy)
print(classification_report(y_test, new_preds))
matriz_confusão = confusion_matrix(y_test, new_preds)
plt.figure(figsize=(8,8))
sns.heatmap(matriz_confusão,annot=True)
```

```python
from sklearn import metrics
auc = metrics.roc_auc_score(y_test, new_preds)

false_positive_rate, true_positive_rate, thresolds = metrics.roc_curve(y_test, new_preds)

plt.figure(figsize=(6, 6), dpi=100)
plt.axis('scaled')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.title("AUC & ROC Curve")
plt.plot(false_positive_rate, true_positive_rate,'--' )
plt.fill_between(false_positive_rate, true_positive_rate, facecolor='lightblue', alpha=0.7)
plt.text(0.95, 0.05, 'AUC = %0.4f' % auc, ha='right', fontsize=12, weight='bold', color='blue')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
```
